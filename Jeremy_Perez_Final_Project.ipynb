```python
#BA final project
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib.ticker import StrMethodFormatter
import seaborn as sns

allregions_total_sales = pd.read_csv("/content/totalsales_EVERYREGION.csv")


allregions_top_sales = allregions_total_sales.head(10).round(2)
allregions_top_sales # shows top 10 items sales from ALL regions
```

```python
#Plot

product_name = allregions_top_sales['Product_name']
total_sales = allregions_top_sales['sum_sales']

plt.figure(figsize=(10,5))
plt.xticks(rotation=45, ha='right')

plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:.0f}'))

plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Top Ten Best Selling Products")
plt.bar(product_name, total_sales)
```

```python
all_low_sales = allregions_total_sales.tail(10).round(2) #shows 10 lowest items sales from ALL regions.
all_low_sales
```

```python
#Plot
names= all_low_sales['Product_name']
value= all_low_sales['sum_sales']
plt.figure(figsize=(10,5))
plt.xticks(rotation=45, ha='right')

plt.gca().yaxis.set_major_formatter(StrMethodFormatter('{x:.0f}'))


plt.xlabel("Products")
plt.ylabel("Total Sales")
plt.title("Ten Lowest Selling Products")
plt.bar(names, value)
```

```python
data = pd.read_csv("/content/testing.csv")
data['Year'] = pd.to_datetime(data['Date_of_Order']).dt.year

# Filter data for the year 2015
data_2015 = data[data['Year'] == 2015]

# Group the data by Product_name and calculate the total Units_sold for each product
product_units_sold_2015 = data_2015.groupby('Product_name')['Units_sold'].sum()

# Sort the products by total units sold in descending order and select the top 10
top_products_2015 = product_units_sold_2015.nlargest(10)
top_products_2015
```

```python
# Plotting
plt.figure(figsize=(10, 6))
top_products_2015.plot(kind='bar', color='limegreen')
plt.xlabel('Product Name')
plt.ylabel('Quantity')
plt.title('Top 10 Best Selling Products in 2015')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

Profits

```python
# Load the data
profit_data = pd.read_csv("/content/profits.csv")

# Converts 'Date_of_order' column to datetime
profit_data['Date_of_order'] = pd.to_datetime(profit_data['Date_of_order'])

# Extract year and month
profit_data['Year'] = profit_data['Date_of_order'].dt.year
profit_data['Month'] = profit_data['Date_of_order'].dt.month

# Group the profit_data by year and month and calculate total profit
monthly_profit = profit_data.groupby(['Year', 'Month'])['Order_Profit_Per_Order'].sum().round(2)
monthly_profit_df=pd.DataFrame(monthly_profit)
monthly_profit_df
```

```python
# Plotting
plt.figure(figsize=(10, 6))

for year in monthly_profit.index.levels[0]:
    year_profit_data = monthly_profit.loc[year]
    plt.plot(year_profit_data.index, year_profit_data.values, label=year,marker='o')

plt.xlabel('Month')
plt.ylabel('Total Profit ($)')
plt.title('Total Profits from 2015 to 2018')
plt.legend(loc='best')
plt.grid()

plt.show()
```
