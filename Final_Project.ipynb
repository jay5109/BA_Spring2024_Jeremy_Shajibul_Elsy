```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file
data = pd.read_csv('/content/Shajibul1.csv')

# Create the graph
plt.figure(figsize=(12, 7))
for region in data['Region'].unique():
    # Filter the data for each region
    region_data = data[data['Region'] == region]
    # Convert Sale_Month to string for better tick formatting on the x-axis
    plt.bar(region_data['Sale_Month'].astype(str), region_data['TotalRegionalSales'], label=region)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Total Regional Sales')
plt.title('Total Sales in 2017')
plt.legend(title='Region')


# Show the graph
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from a CSV file
data = pd.read_csv('/content/Shajibul2.csv')

# Create the graph
plt.figure(figsize=(3, 7))
for region in data['Region'].unique():
    # Filter the data for each region
    region_data = data[data['Region'] == region]
    # Plotting total sales for each region by month
    plt.bar(region_data['Sale_Month'].astype(str), region_data['TotalRegionalSales'], label=region, width=0.04)

# Add labels and title
plt.xlabel('Month')
plt.ylabel('Total Regional Sales')
plt.title('Total Sales in 2018')
plt.legend(title='Region')


# Show the graph
plt.show()
```
