```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/(Shajibul)Results.csv')




plt.figure(figsize=(14, 8))

highest_spenders = df[df['Category'] == 'Highest Spenders']
plt.bar(highest_spenders['Customer_Fname'] + " " + highest_spenders['Customer_Lname'], highest_spenders['TotalSpent'], color='blue')
plt.title('Total Spent by Highest Spenders')
plt.xlabel('Customer Name')
plt.ylabel('Total Spent')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(14, 8))

lowest_spenders = df[df['Category'] == 'Lowest Spenders']
plt.bar(lowest_spenders['Customer_Fname'] + " " + lowest_spenders['Customer_Lname'], lowest_spenders['TotalSpent'], color='red')
plt.title('Total Spent by Lowest Spenders')
plt.xlabel('Customer Name')
plt.ylabel('Total Spent')
plt.xticks(rotation=90)
plt.show()




```

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv('/content/Results2.csv')


plt.figure(figsize=(20, 10))


most_frequent = df[df['Category'] == 'Most Frequent Buyers']


plt.bar(most_frequent['Customer_Fname'] + " " + most_frequent['Customer_Lname'], most_frequent['TotalOrders'], color='blue', label='Total Orders')
plt.title('Total Orders for Most Frequent Buyers')
plt.xlabel('Customer Name')
plt.ylabel('Total Orders')
plt.xticks(rotation=90)
plt.legend()
plt.show()


plt.figure(figsize=(20, 10))


least_frequent = df[df['Category'] == 'Least Frequent Buyers']


plt.bar(least_frequent['Customer_Fname'] + " " + least_frequent['Customer_Lname'], least_frequent['TotalOrders'], color='red', label='Total Orders')
plt.title('Total Orders for Least Frequent Buyers')
plt.xlabel('Customer Name')
plt.ylabel('Total Orders')
plt.xticks(rotation=90)
plt.legend()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/Results3.csv')


highest_sales = df[df['Category'] == 'Highest Sales']


lowest_sales = df[df['Category'] == 'Lowest Sales']


plt.figure(figsize=(14, 7))
plt.bar(highest_sales['Customer_State'], highest_sales['TotalRegionalSales'], color='green')
plt.title('Total Regional Sales for Highest Sales by State')
plt.xlabel('State')
plt.ylabel('Total Regional Sales')
plt.xticks(rotation=90)
plt.show()


plt.figure(figsize=(14, 7))
plt.bar(lowest_sales['Customer_State'], lowest_sales['TotalRegionalSales'], color='red')
plt.title('Total Regional Sales for Lowest Sales by State')
plt.xlabel('State')
plt.ylabel('Total Regional Sales')
plt.xticks(rotation=90)
plt.show()
```
